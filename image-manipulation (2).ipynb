```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/kaggle/input/datasets/oddyvirgantara/psd-dataset/psd_dataset/lake.jpg")

img_cvt = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_cvt)
plt.show()
```

```python
print(img_cvt.shape)
```

```python
img_hujan = cv.imread("/kaggle/input/datasets/oddyvirgantara/psd-dataset/psd_dataset/rain.jpg")
img_hujan = cv.cvtColor(img_hujan, cv.COLOR_BGR2RGB)

plt.imshow(img_hujan)
plt.show()
```

```python
print(img_hujan.shape)
```

```python
img_danau = cv.resize(img_cvt, (600, 400))
img_hujan = cv.resize(img_hujan, (600, 400))

img_blend = cv.addWeighted(img_danau, 1.5, img_hujan, 0.5, 0)
plt.figure(figsize=(20,20))
plt.subplot(1, 3, 1)
plt.imshow(img_cvt)

plt.subplot(1, 3, 2)
plt.imshow(img_hujan)

plt.subplot(1, 3, 3)
plt.imshow(img_blend)
plt.show()
```

Image Negatif

```python
img_danau_negatif = 255 - img_danau

plt.figure(figsize=(20,20))
plt.subplot(1, 2, 1)
plt.imshow(img_cvt)

plt.subplot(1, 2, 2)
plt.imshow(img_danau_negatif)


plt.show()
```
